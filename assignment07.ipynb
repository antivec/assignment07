# Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Given datas for polynomial curve fitting

```python
num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 
```

```python
def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f
```

```python
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std
```

# Plot clean data

```python
plt.plot(x, y1, 'k.')
plt.show()
```

# Plot noisy data

```python
plt.plot(x, y2, 'k.')
plt.show()
```

# Plot the polynomial curves varying p = 0 ~ 9

```python
for p in range(0,10):
    z4 = np.polyfit(x,y2,p)
    p4 = np.poly1d(z4)
    print("p = ", p)
    #print("p = " , p, '\n' , " curve function = ",p4)
    plt.plot(x, p4(x),'-g')
    plt.show()
```

# Plot the fitError with varying p
\begin{equation*}
\ error= \sum_{j=1}^n (y_j - \overline{X_j})^2
\end{equation*}


```python
for p in range(0,10):
    pf = np.polyfit(x,y2,p)
    p1 = np.poly1d(pf)
    f = np.polyval(pf,x)
    fitError = (y2-f) * (y2-f)
    plt.plot(x, fitError,'-g')
    print("p = ", p)
    #print("fitError = " , y2-f)
    plt.show()
```
